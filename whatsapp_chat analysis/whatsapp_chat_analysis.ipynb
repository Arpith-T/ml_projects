#### STEP1: create 2 tables `date` and `message` (this contains both user name and message)
* The above can be done via `regular expression`
> Refer - https://regex101.com/

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)
messages
```

```python
type(messages)
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
len(messages)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date': dates})
```

```python
df["message_date"]
```

```python
# convert message_date type

df['message_date'] = pd.to_datetime(df['message_date'], format="%d/%m/%y, %H:%M - ")

df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

> Refer: https://docs.python.org/3/library/datetime.html

```python
df.shape
```

### Seperate `user` from the `message`

```python
users = []
messages = []
for message in df["user_message"]:
    entry = re.split("([\w\W]+?):\s", message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append("group notification")
        messages.append(entry[0])

df["users"] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```
